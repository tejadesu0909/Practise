```python
# How do I sort the series or a data frame. 

movies = pd.read_csv("http://bit.ly/imdbratings")
movies
```

```python
movies.title.sort_values().tail(20)
```

```python
movies.sort_values(by="title").head(20)
```

```python
movies[movies["duration"]>=200]
```

```python
movies.loc[movies["duration"]>=200]
```

```python
# Using AND OR conditions to filter the rows. 

movies.loc[(movies.duration>=200) & (movies.genre == "Crime")]
```

```python
# If we want to filter same column with different values such as genre has crime or adventure or drama.
# We can use by giving conditions individually, 

movies.loc[(movies.genre == "Drama" )| (movies.genre == "Crime" ) | (movies.genre == "Adventure" )]
```

```python
# Or we can write it in optimistically by writing, 

movies.loc[movies.genre.isin(["Crime", "Drama", "Adventure"])]
```
